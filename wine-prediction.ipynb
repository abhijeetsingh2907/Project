```python
num_attributes=wine.select_dtypes(include="float").columns
num_attributes
```

```python
good_df=wine[wine["quality"]==1]
bad_df=wine[wine["quality"]==0]
```

```python
from scipy import stats
```

```python
statistical_significance=[]
for attr in num_attributes:
    statistic, p=stats.f_oneway(good_df[attr].values,bad_df[attr].values)
    statistical_significance.append([attr,round(p,6)])
statistical_significance=pd.DataFrame(statistical_significance)
statistical_significance.columns=["Attribute","P-value"]
print(statistical_significance)
```

```python
cat_attributes=["type"]
```

```python
statistical_significance=[]
for attr in cat_attributes:
    data_count=pd.crosstab(wine[attr],wine["quality"]).reset_index()
    obs=np.asarray(data_count[[0,1]])
    chi2, p, dof, expected = stats.chi2_contingency(obs)
    statistical_significance.append([attr,round(p,6)])
statistical_significance=pd.DataFrame(statistical_significance)
statistical_significance.columns=["Attribute","P-value"]
print(statistical_significance)
```

```python
x = wine.drop(['quality',"pH"], axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x)
X_standard = scaler.transform(x)
```

```python
X = X_standard
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = .25, random_state = 113)
```

## Using CLassification Models

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score,classification_report,confusion_matrix
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
rfc_test_pred = rfc.predict(X_test)
```

```python
print("Test:=",accuracy_score(rfc_test_pred, Y_test))
```

```python
print(classification_report(rfc_test_pred, Y_test))
```

```python
print(confusion_matrix(rfc_test_pred, Y_test))
```
